```python
def create_inout_sequences(input_data, tw, pw):
    inout_seq = []
    train_seq = []
    train_label =[]
    all_train_seqs = []
    all_train_labels = []
    L = len(input_data)
    if L == tw+pw:offset = 1
    else: offset = 0
    
    for i in range(0,L-tw-pw+offset):
        train_seq = input_data[i:i+tw]/input_data[i]
        train_label = input_data[i+tw:i+tw+pw,1]/ input_data[i,1]
        if len(train_seq) != tw:continue
        
        inout_seq.append((train_seq ,train_label))
    return inout_seq
```

```python
n_features = 8 # this is number of parallel inputs
n_timesteps = 10 # this is number of timesteps
```

```python
class MV_LSTM(torch.nn.Module):
    def __init__(self,n_features,seq_length):
        super(MV_LSTM, self).__init__()
        self.n_features = n_features
        self.seq_len = seq_length
        self.n_hidden = 200 # number of hidden states
        self.n_layers = 2 # number of LSTM layers (stacked)
    
        self.l_lstm = torch.nn.LSTM(input_size = n_features, 
                                 hidden_size = self.n_hidden,
                                 num_layers = self.n_layers, 
                                 batch_first = True)
        self.l_lstm2 = torch.nn.LSTM(input_size = self.n_hidden, 
                                 hidden_size = self.n_hidden,
                                 num_layers = self.n_layers, 
                                 batch_first = True)
        # according to pytorch docs LSTM output is 
        # (batch_size,seq_len, num_directions * hidden_size)
        # when considering batch_first = True
        self.l_linear = torch.nn.Linear(self.n_hidden*self.seq_len, 5)
        
    
    def init_hidden(self, batch_size):
        # even with batch_first = True this remains same as docs
        hidden_state = torch.zeros(self.n_layers,batch_size,self.n_hidden).cuda()
        cell_state = torch.zeros(self.n_layers,batch_size,self.n_hidden).cuda()
        self.hidden = (hidden_state, cell_state)
        
        hidden_state2 = torch.zeros(self.n_layers,batch_size,self.n_hidden).cuda()
        cell_state2 = torch.zeros(self.n_layers,batch_size,self.n_hidden).cuda()
        self.hidden2 = (hidden_state2, cell_state2)
    
    
    def forward(self, x):        
        batch_size, seq_len, _ = x.size()
        lstm_out, self.hidden = self.l_lstm(x)
        lstm_out2, self.hidden2 = self.l_lstm2(lstm_out)
        # lstm_out(with batch_first = True) is 
        # (batch_size,seq_len,num_directions * hidden_size)
        # for following linear layer we want to keep batch_size dimension and merge rest       
        # .contiguous() -> solves tensor compatibility error
        x = lstm_out2.contiguous().view(batch_size,-1)
        return self.l_linear(x)
```

```python
# create NN
mv_net = MV_LSTM(n_features,n_timesteps)
criterion = torch.nn.L1Loss() # reduction='sum' created huge loss value
optimizer = torch.optim.Adam(mv_net.parameters(), lr=.001)
print(mv_net)

batch_size = 256
```

```python
# Data Loader for easy mini-batch return in training
train_loader = torch.utils.data.DataLoader(dataset=all_train_seq,batch_size=batch_size, shuffle=True)
```

```python
EPOCH = 25
#mv_net.load_state_dict(torch.load(r'C:\Users\conno\Desktop\PyTorch\stockmodel3.pth'))
mv_net.train()
mv_net = mv_net.cuda()
decayRate = 0.96

my_lr_scheduler = torch.optim.lr_scheduler.ExponentialLR(optimizer=optimizer, gamma=decayRate)
for epoch in range(EPOCH):
    running_loss = 0
    for step, (x, y) in enumerate(train_loader):
        x_batch = x.to(device)
        y_batch = y.to(device)
        mv_net.init_hidden(x_batch.size(0))
        output = mv_net(x_batch)
        loss = criterion(output, y_batch)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        my_lr_scheduler.step()
        running_loss += loss.item()
    print('[%d, %5d] running_loss: %.3f'  %(epoch+1, step+1, running_loss/(step+1)))
    
```
